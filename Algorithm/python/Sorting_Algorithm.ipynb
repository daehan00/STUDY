# data sorting algorithm

## 자료 선택 시 고려사항
### 자료의 양
시간복잡도 : 양이 많은 경우 쉘 정렬 사용, 적은 경우 삽입정렬
### 사용 가능한 기억 공간의 크기
기억공간 사용량 적은 것 선택
### 정렬을 위한 자료 이동 빈도 수

```python
def selectionSort(List):
    for i in range(len(List)-1):
        min_position = i
        for j in range(i+1,len(List)):
            if List[j] < List[min_position]:
                min_position = j
        if min_position != i:
            tmp = List[i]
            List[i] = List[min_position]
            List[min_position] = tmp
```

```python
sample_list = [5,2,3,4,11,7,1,9,10]
selectionSort(sample_list)
print(sample_list)
```

```python
def insertionSort(List):
    for i in range(1,len(List)):
        position = i
        current_value = List[i]
        while position>0 and List[position-1]>current_value:
            List[position] = List[position-1]
            position = position-1
        if position != i:
            List[position] = current_value
```

```python
sample_list = [5,2,3,4,11,7,1,9,10]
insertionSort(sample_list)
print(sample_list)
```

```python
def i_MergeSort(lst):
    lists = [[x] for x in lst]
    while len(lists)>1:
        lists = merge_lists(lists)
    return lists[0]

def merge_lists(lists):
    result = []
    for i in range(0, len(lists)//2):
        result.append(merge2(lists[i*2], lists[i*2+1]))
    if len(lists) % 2:
        result.append(lists[-1])
    return result

def merge2(xs, ys):
    i, j = 0, 0
    result = []
    while i < len(xs) and j < len(ys):
        x, y = xs[i], ys[j]
        if x > y:
            result.append(y)
            j += 1
        else:
            result.append(x)
            i += 1
        result.extend(xs[i:])
        result.extend(ys[j:])
        return result
```

```python
sample_list = [5,2,3,4,11,7,1,9,10]
sorted_list = i_MergeSort(sample_list)
print(sorted_list)
```

```python
def shellSort(List):
    n = len(List)
    gap = n//2
    while gap > 0:
        for i in range(gap, n):
            tmp = List[i]
            j = i
            while j >= gap and List[j-gap] > tmp:
                List[j] = List[j-gap]
                j -= gap
            List[j] = tmp
        gap //= 2
    return List
```

```python
sample_list = [5,2,3,4,11,7,1,9,10]
shellSort(sample_list)
print(sample_list)
```
